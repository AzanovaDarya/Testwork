### 1. Построение модели перевозок от поставщиков к потребителям при минимизации издержек на транспортировку

```python
from pulp import *
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def RouteCost(C, b, a):
    """
    C (np.array): стоимость перевозки
    b (np.array): потребность потребителей
    a (np.array): объемы складов
    """
    
    n, m = C.shape
    x = LpVariable.dicts("x",[(i,j) for i in range(1,n + 1) for j in range(1, m + 1)],lowBound=0, upBound=None, cat='Integer')

    problem = pulp.LpProblem('0',LpMinimize)

    problem += lpSum(x[i + 1,j + 1] * C[i,j] for i in range(n) for j in range(m)) #Функция цели

    for i in range(1, n + 1):
        problem +=lpSum(x[i, j] for j in range(1, m + 1)) <= a[i-1] #склад
    
    for j in range(1, m + 1):
        problem +=lpSum(x[i, j] for i in range(1, n + 1)) == b[j-1] #клиент

    problem.solve()
    print ("Результат:")
    for variable in problem.variables():
        print (variable.name, "=", variable.varValue)
        
    print("Model Status:{}".format(LpStatus[problem.status]))
    o = [{'Name':name,'Constraint':c,'shadow price':c.pi,'slack': c.slack} for name, c in problem.constraints.items()]

    print(pd.DataFrame(o))
    print ("\nСтоимость доставки: ", value(problem.objective))
    return abs(value(problem.objective))
```

```python
C = np.array([[3, 20, 8, 13, 4, 100], [4, 4, 18, 14, 3, 0], [10, 4, 18,8, 6, 0], [7, 19, 17, 10, 1, 100]]) #стоимость перевозки
b = np.array([10, 30, 40, 50, 70, 30]) # потребность потребителей
a= np.array([80, 60, 30, 60]) # объемы складов

RouteCost(C = C, b = b, a = a)
```

## 2. Построение плана перевозок от поставщиков к потребителям с учетом доступного транспорта при минимизации издержек на транспортировку

```python
def RouteCost(cost, ability, quantity, consumers, warehouses):
    """
    cost (np.array): стоимость перевозки
    ability (np.array): грузоподъемность транспорта
    quantity (np.array): количество транспорта, доступное для перевозок от данного поставщика
    consumers (np.array): потребность потребителей
    warehouses (np.array): объемы складов
    """
    
    n, _ = route_cost.shape
    
    x = LpVariable.dicts("x",[(i) for i in range(1, n + 1)],lowBound=0, upBound=None, cat='Integer')

    problem = pulp.LpProblem('0',LpMinimize)

    problem += lpSum(x[i+1] * cost[i] for i in range(n)) #Функция цели

    for i in range(12):
        problem += lpSum(x[i + 1] * ability[i]) <= warehouses[0] #склад1

    for i in range(12, n):
        problem += lpSum(x[i + 1] * ability[i]) <= warehouses[1] #склад2
    
    problem += x[1] * ability[0] + x[2] * ability[1] + x[3] * ability[2] + x[4] * ability[3] + x[13] * ability[12] + x[14] * ability[13] + x[15] * ability[14] + x[16] * ability[15] == consumers[0] #клиент1

    problem += x[5] * ability[4] + x[6] * ability[5] + x[7] * ability[6] + x[8] * ability[7] + x[17] * ability[16] + x[18] * ability[17] + x[19] * ability[18] + x[20] * ability[19] == consumers[1] #клиент2
                        
    problem += x[9] * ability[8] + x[10] * ability[9] + x[11] * ability[10] + x[12] * ability[11] + x[21] * ability[20] + x[22] * ability[21] + x[23] * ability[22] == consumers[2] #клиент3
                        
    problem += x[1] + x[5] + x[9] <= quantity[0] #количество транспорта в Москве вместимостью 3 тонны
    
    problem += x[2] + x[6] + x[10] <= quantity[1] #количество транспорта в Москве вместимостью 5 тонн
                        
    problem += x[3] + x[7] + x[11] <= quantity[2] #количество транспорта в Москве вместимостью 10 тонн
                        
    problem += x[4] + x[8] + x[12] <= quantity[3] #количество транспорта в Москве вместимостью 20 тонн
                        
    problem += x[13]  + x[17] + x[21] <= quantity[4] #количество транспорта в Ярославле вместимостью 3 тонны
                        
    problem += x[14] + x[18] + x[22] <= quantity[5] #количество транспорта в Ярославле вместимостью 5 тонн                    
    
    problem += x[15] + x[19] + x[23] <= quantity[6] #количество транспорта в Ярославле вместимостью 10 тонн
                        
    problem += x[16] + x[20] <= quantity[7]          #количество транспорта в Ярославле вместимостью 20 тонн                    
    
                           
    problem.solve()
    print(problem)
    print ("Результат:")
    for variable in problem.variables():
        print (variable.name, "=", variable.varValue)
        
    print("Model Status:{}".format(LpStatus[problem.status]))
    o = [{'Name':name,'Constraint':c,'shadow price':c.pi,'slack': c.slack} for name, c in problem.constraints.items()]

    print(pd.DataFrame(o))
    print ("\nСтоимость доставки: ", value(problem.objective))
    return abs(value(problem.objective))
```

# Тестовое задание №3 
<br>

<br><br>
&nbsp;&nbsp;&nbsp;&nbsp; **Описание работы:**
&nbsp;&nbsp;&nbsp;&nbsp;
<br><br>
&nbsp;&nbsp;&nbsp;&nbsp; Имеется набор узлов (тбл nodes), часть из которых узлы погрузки, другие выгрузки, разница указывается в поле unl (нет погрузки, да выгрузки).
Необходимо вычислить потоки из узлов погрузки в узлы выгрузки (груженые) и из узлов выгрузки в узлы погрузки (порожние). Возможные направления груженых потоков зафиксированы в тбл fr_arcs, порожний поток может быть направлен из любого узла выгрузки в любой узел погрузки. 
Груженые потоки приносят доход, его величина на единицу потока указана в поле Comission в тбл fr_arcs. Порожние потоки приводят к расходу, величина расхода на единицу потока указана в поле trf таблицы emp_trfs. 

&nbsp;&nbsp;&nbsp;&nbsp;Задача состоит в нахождении груженых и порожних потоков, при которых величина прибыли, определяемая как разность суммарного дохода и суммарного расхода, была бы максимальна.

&nbsp;&nbsp;&nbsp;&nbsp;При этом необходимо соблюдение следующих условий:
1. Потоки неотрицательны.
2. Во всех узлах соблюдается баланс: входящий поток равен выходящему.
3. Груженый поток не может превзойти установленный предел (поле Volume_limit в тбл fr_arcs). 
4. Для некоторых из груженых направлений груженый поток должен быть точно равен установленному пределу, такие направления указываются значением да в поле Obligatory в тбл fr_arcs.
5. Взвешенный полный суммарный поток (груженый и порожний в совокупности) не должен превосходить 3237980,35. Веса указаны в поле weight тбл fr_arcs (для груженых) и emp_trfs (для порожних).

&nbsp;&nbsp;&nbsp;&nbsp;Груженые потоки отразить в поле Volume_fact тбл fr_arcs, порожние потоки в тбл emp_arcs.

## 3.2. Построение модели.

```python
n, _ = fr_arcs.shape
m, _ = emp_trfs.shape
```

```python
#create variables for load nodes
load = LpVariable.dicts("load", [(fr_arcs['loa'].loc[i], fr_arcs['unl'].loc[i], fr_arcs['volume_limit'].loc[i]) for i in range(n)],lowBound=0, upBound=None, cat='Integer')
len(load)
```

```python
#create variables for unload nodes
unload = LpVariable.dicts("unload", [(emp_trfs['unl'].loc[i], emp_trfs['loa'].loc[i]) for i in range(m)],lowBound=0, upBound=None, cat='Integer')
len(unload)
```

```python
# Create problem function
problem = pulp.LpProblem('VolumeNodes',LpMaximize)

#Функция цели суммарный доход:
problem += lpSum((
    load[fr_arcs['loa'].loc[i], 
         fr_arcs['unl'].loc[i], 
         fr_arcs['volume_limit'].loc[i]]) * 
    fr_arcs['comission'].loc[i]  for i in range(n)) 

#Функция цели суммарный расход:
problem += lpSum(- (
    unload[emp_trfs['unl'].loc[j], 
           emp_trfs['loa'].loc[j]]) * 
                 emp_trfs['trf'].loc[j] for j in range(m)) 

# Баланс в узлах: входящий поток равен выходящему:
for i in range(n):
    problem += ((load[fr_arcs['loa'].loc[i], 
                 fr_arcs['unl'].loc[i], 
                 fr_arcs['volume_limit'].loc[i]]) - (unload[fr_arcs['unl'].loc[i], 
                                                           fr_arcs['loa'].loc[i]]) == 0)  

 # Груженый поток не превосходит установленный предел:
for i in range(n):
    problem += (load[fr_arcs['loa'].loc[i], 
                fr_arcs['unl'].loc[i], 
                fr_arcs['volume_limit'].loc[i]] - fr_arcs['volume_limit'].loc[i] <= 0)

# Груженый поток точно равен установленному пределу для узлов Obligatory == True:
for i in range(n):
    if fr_arcs['obligatory'].loc[i] == True:
        problem += (load[fr_arcs['loa'].loc[i], 
                        fr_arcs['unl'].loc[i], 
                        fr_arcs['volume_limit'].loc[i]] - fr_arcs['volume_limit'].loc[i] == 0)

# Взвешенный полный суммарный поток (груженый и порожний в совокупности) не должен превосходить 3237980,35
load_weight = lpSum(
    (load[fr_arcs['loa'].loc[i], 
          fr_arcs['unl'].loc[i], 
          fr_arcs['volume_limit'].loc[i]]) * fr_arcs['weight'].loc[i] for i in range(n))

unload_weigth = lpSum((unload[emp_trfs['unl'].loc[j], 
                              emp_trfs['loa'].loc[j]]) * emp_trfs['weight'].loc[j] for j in range(m))


problem += (load_weight + unload_weigth) <= 3237980.35
```

```python
problem.solve()
print ("\nМаксимальная прибыль: ", value(problem.objective))
```
