```python
#Импорт библиотек

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set2')
import datetime as dt


#Выгрузка данных из файлов csv

courses = pd.read_csv('courses.csv', index_col = 0)

students = pd.read_csv('students.csv', index_col = 'id_', parse_dates = ['birthday'], dayfirst = False)

course_contents = pd.read_csv('course_contents.csv')

progresses = pd.read_csv('progresses.csv')

progress_phases = pd.read_csv('progress_phases.csv', parse_dates = ['start_date', 'finish_date'], dayfirst = False)


# Объединение данных из progress_phases, progresses и courses в один датасет

df = progress_phases.merge(progresses, how = 'left', left_on = 'progress_id', right_on = 'id').drop(columns =  'id')
df = df.merge(courses, how = 'left', left_on = 'course_id', right_on = 'id').drop(columns =  'id')

# Переименование заголовков столбцов
df.columns = ['progress_id',
 'module_number',
 'lesson_number',
 'status',
 'start_date',
 'finish_date',
 'student_id',
 'course_id',
 'course_title',
 'course_field']

# Добавление в объединенный датасет данных из course_contents
main = df.merge(course_contents[['course_id',
                               'module_number',
                               'lesson_number',
                               'module_title', 
                               'lesson_title', 
                               'lesson_token', 
                               'is_video', 
                               'is_homework']], how = 'left', on = ['course_id', 'module_number','lesson_number'])
main
```

```python
# проверка основного датасета
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'] 
```

```python
# Рассчет общего количества курсов в датасете

print('Общее количество курсов в датасете: ', main['course_id'].nunique())
```

```python
# Количество модулей на каждом курсе

main.groupby(['course_title', 'course_field'])['module_number'].nunique().sort_values(ascending = False)
```

```python
#количество уроков в каждом модуле на каждом курсе

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(main.groupby(['course_title', 'module_number'])['lesson_number'].nunique())
```

```python
#медианное количество уроков в модуле на каждом курсе

main.groupby(['course_title', 'module_number'], as_index = False)['lesson_number'].nunique().groupby('course_title').\
    agg({'lesson_number' : ['median']})
```

```python
# количество учеников на каждом курсе

main.groupby('course_title', as_index = False)['student_id'].nunique()
```

```python
# минимальный, максимальный, средний, медианный возраст студентов

# заполняем нулевыми значениями те позиции, где отсутствуют данные по дням рождения студента
students.fillna({'birthday' : 0}, inplace = True)

# добавляем данные по дням рождения к основному df в новый df_students
df_students = main.merge(students, how = 'left', left_on = 'student_id', right_on = 'id').drop(columns =  'id')

# отсекаем только те данные, в которых есть дни рождения студентов
df_students = df_students.loc[df_students['birthday'] != 0]

# находим возраст студентов на момент старта курса
df_students['age'] = (pd.to_datetime(df_students['start_date'], dayfirst = False).dt.year 
                      - pd.to_datetime(df_students['birthday'], dayfirst = False, 
                                       errors = 'coerce').dt.year)

# находим основные характеристики столбца ['age']
df_students['age'].describe()
```

```python
# отсекаем данные с ошибками

df_students = df_students.loc[(df_students['age'] > 10) & (df_students['age'] < 100) ]

# находим основные характеристики столбца ['age']
df_students['age'].describe()
```

```python
# проверим возраст после 90 лет

df_students[df_students['age'] > 90]
```

```python
# Поскольку дата рождения 1 января 1920 года, то скорее всего студент поленился заполнять 
# графу дня рождения, эти данные также уберем из df

df_students = df_students.loc[df_students['age'] < 99 ]

# находим основные характеристики столбца ['age']
df_students['age'].describe()
```

```python
# сортируем по возрасту студента и фильтруем df по уникальным id студентов

unique_students = df_students.sort_values('age').drop_duplicates(subset = 'student_id')
unique_students
```

```python
# находим основные характеристики столбца ['age']
unique_students['age'].describe()
# Min возраст студентов 12 лет, max = 89 лет, средний = 29 лет, медианный = 28 лет.
```

```python
# Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
df_age_course = df_students.groupby('course_title').agg({'age':['min', 'max', 'mean', 'median']})
df_age_course
```

```python
# Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

df_students_count = df_students.groupby('course_title')['student_id'].nunique().sort_values(ascending = False).reset_index()

fig, ax = plt.subplots(figsize = (12, 6))
ax.bar(df_students_count['course_title'], df_students_count['student_id'])      
ax.set_title('Количество студентов на курсе')
ax.set_xlabel('Курс')
ax.set_ylabel('Количество человек')
ax.tick_params(axis = 'x', rotation = 90)
plt.show()
```

```python
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов 
#на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать 
#информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
#На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно 
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np



df_students_ucount = df_students.groupby(['course_title', 'course_field'])['student_id'].nunique()\
            .sort_values(ascending = False).reset_index()





fig, ax = plt.subplots(figsize = (12, 6))
sns.barplot(
    data = df_students_ucount , 
    x = 'student_id', 
    y = 'course_title', 
    ax = ax,
    ci = None, 
    alpha = 0.1, 
    hue = 'course_field',
    saturation = 200,
    dodge = False
)
ax.axvline(
    x = df_students_ucount['student_id'].median(), 
    color = 'blue', 
    linewidth = 1, 
    linestyle = '--'
)
ax.set_title('Количество уникальных студентов на курсе')
ax.set_xlabel('Количество человек')
ax.legend()
plt.show()

figures = []
figures.append(fig)
# Массив фигур.
# figures = []

# Создание файла и сохранение каждой фигуры.
pdf = PdfPages("Figures.pdf")
for figure in figures:
    pdf.savefig(figure)


# Сохранение файла
pdf.close()
```

```python
#Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, 
#приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), 
#к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).


df_students = main.merge(students, how = 'left', left_on = 'student_id', right_on = 'id').drop(columns =  'id')
#Считаем количество начавших ДЗ студентов в модуле
start_count = df_students[(
    df_students['is_homework'] == True)][['course_title',
                                          'progress_id',
                                          'module_number',
                                          'start_date']].copy()

start_count['start_date'] = pd.to_datetime(start_count.start_date, utc = True)

start_count = start_count.groupby(
    ['course_title', 'module_number', 'progress_id'], as_index = False).agg({'start_date': lambda x: min(x)})

start_count = start_count.groupby(['course_title','module_number'])['start_date'].count().unstack(fill_value=0).stack().to_frame().reset_index()

start_count.columns = ['course_title','module_number', 'start_count']

#Считаем количество закончивших ДЗ студентов в модуле
finish_count = df_students[(
    df_students['is_homework'] == True) & (df_students['status'] == 'done')][['course_title',
                                                                              'progress_id',
                                                                              'module_number',
                                                                              'finish_date']].copy()

finish_count['finish_date'] = pd.to_datetime(finish_count.finish_date, utc = True)

finish_count = finish_count.groupby(
    ['course_title', 'module_number', 'progress_id'], as_index = False).agg({'finish_date': lambda x: max(x)})

finish_count = finish_count.groupby(['course_title','module_number'])['finish_date'].count().unstack(fill_value=0).stack().to_frame().reset_index()

finish_count.columns = ['course_title','module_number', 'finish_count']
finish_count['finish_count'] = finish_count.groupby(['course_title'])['finish_count'].shift(1)

#Соединяем датафреймы
df_conversion = start_count.merge(finish_count, how = 'left', on = ['course_title', 'module_number'])

#Считаем конверсию
df_conversion['conversion'] = df_conversion[['start_count','finish_count']].apply(lambda x: 1 if pd.isna(x['finish_count']) else x['start_count'] / x['finish_count'],axis=1)

df_conversion

```
